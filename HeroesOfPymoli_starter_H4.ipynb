```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
#Display the total number of players
total_players=purchase_data ["SN"].nunique()
total_players = pd.DataFrame({"Total players": [total_players]})
total_players
```

```python
#Obtain total number of unique items
number_unique=purchase_data ["Item Name"].nunique()
number_unique
```

```python
#Obtain average price of total purchases
average_price=round(purchase_data["Price"].sum() / purchase_data["Price"].count(),2)
average_price
```

```python
#Obtain total of purchases
number_purchases=purchase_data["Price"].count()
number_purchases
```

```python
#Obtain total of revenue
total_revenue=purchase_data["Price"].sum()
total_revenue
```

```python
#Summary DataFrame
purchasing_analysis=pd.DataFrame({"Number of Unique Items": [number_unique],
                         "Average Price": "$" + str(average_price),
                         "Number of Purchases": [number_purchases],
                                  "Total Revenue": "$" + str(total_revenue)})
purchasing_analysis
```

```python
#Gender demographics - Count of each one
count_tgender=purchase_data.groupby(["Gender"])
gender_demographic = count_tgender["SN"].nunique()
Total_Gender=pd.DataFrame(gender_demographic)
Total_Gender.columns=["Total Count"]
Total_or = Total_Gender.sort_values("Total Count", ascending = False)
Total_or
```

```python
#Gender demographics - Percentage of each one
percentage_players=round(((gender_demographic/purchase_data["SN"].nunique())*100),2)
Percentage_Gender=pd.DataFrame(percentage_players)
Percentage_Gender.columns=["Percentage of Players"]
Percentage_Gender
```

```python
Gender = Total_Gender.copy() 
Gender["Percentage of Players"] = Percentage_Gender["Percentage of Players"]
Gender_or = Gender.sort_values("Total Count",ascending=False)
Gender_or = Gender_or.reset_index()
Gender_or
```

```python
#Purchasing analysis (Gender) - Total purchase
count_gender=purchase_data.groupby(["Gender"])
gender_purchase = count_gender["Purchase ID"].nunique()
Total_purchase_g=pd.DataFrame(gender_purchase)
Total_purchase_g.columns = ["Purchase Count"]
Total_purchase_g
```

```python
#Purchasing analysis (Gender) - average purchase
count_gender=purchase_data.groupby(["Gender"])
gender_purchase = round((count_gender["Price"].sum())/count_gender["Price"].count(),2)
avg_purchase_g=pd.DataFrame(gender_purchase)
avg_purchase_g.columns = ["Average Purchase Price"]
avg_purchase_g
```

```python
#Purchasing analysis (Gender) - total purchase
count_gender=purchase_data.groupby(["Gender"])
gender_purchase = round((count_gender["Price"].sum()),2)
sum_purchase_g=pd.DataFrame(gender_purchase)
sum_purchase_g.columns = ["Total Purchase Value"]
sum_purchase_g
```

```python
#Purchasing analysis (Gender) - avg purchase per person
count_gender=purchase_data.groupby(["Gender"])
gender_purchase = round((count_gender["Price"].sum())/count_gender["SN"].nunique(),2)
avg_pp_g=pd.DataFrame(gender_purchase)
avg_pp_g.columns = ["Avg Total Purchase per Person"]
avg_pp_g
```

```python
#Purchasing analysis (Gender)
Gender_purchase_a = Total_purchase_g.copy() 
Gender_purchase_a["Average Purchase Price"] = avg_purchase_g["Average Purchase Price"]
Gender_purchase_a["Total Purchase Value"] = sum_purchase_g["Total Purchase Value"]
Gender_purchase_a["Avg Total Purchase per Person"] = avg_pp_g["Avg Total Purchase per Person"]                                              
Gender_purchase_a = Gender_purchase_a.reset_index()
Gender_purchase_a
```

```python
#Determine min and max to know our bins limits (Age Demographics)
print(purchase_data["Age"].max())
print(purchase_data["Age"].min())
```

```python
#Create bins for age
bins = [0,9,14,19,24,29,34,39,45]

#Create Labels for bins
group_labels= ("<10","10-14","15-19","20-24","25-29","30-34","35-39","40+")
```

```python
#Define bins and labels base in a column selected
age_group=pd.cut(purchase_data["Age"],bins, labels=group_labels)
age_group.head()
```

```python
#Define table of total of players associate to ranges base in bins already stablish
age_df=purchase_data.groupby(age_group)
age_count=age_df["SN"].nunique()
age_range=pd.DataFrame(age_count)
age_range.columns=["Total count"]
age_range
```

```python
#Define table of total of players associate to ranges base in bins already stablish
age_group_df_percentage=round(((age_df["SN"].nunique()/purchase_data["SN"].nunique())*100),2)
percentage_range=pd.DataFrame(age_group_df_percentage)
percentage_range.columns=["Percentage of Players"]
percentage_range
```

```python
#Create summary age demographics
age_demographics = age_range.copy()
age_demographics["Percentage of Players"] = percentage_range["Percentage of Players"]
age_demographics
```

```python
#Purchasing analysis Age
age_group=pd.cut(purchase_data["Age"],bins, labels=group_labels)
age_group
```

```python
#purchase count general
age=purchase_data.groupby(age_group)
age_count_general=age["Purchase ID"].count()
age_range_general=pd.DataFrame(age_count_general)
age_range_general.columns=["Purchase count"]
age_range_general
```

```python
#purchase average general
age=purchase_data.groupby(age_group)
avg_general_age=round((age["Price"].sum())/age["Purchase ID"].count(),2)
avg_general_age=pd.DataFrame(avg_general_age)
avg_general_age.columns=["Average Purchase Price"]
avg_general_age
```

```python
#purchase value general
age=purchase_data.groupby(age_group)
sum_general_age=round((age["Price"].sum()),2)
sum_general_age=pd.DataFrame(sum_general_age)
sum_general_age.columns=["Total Purchase Value"]
sum_general_age
```

```python
#average total per person general
age=purchase_data.groupby(age_group)
avg_general_pp_age=round((age["Price"].sum())/age["SN"].nunique(),2)
avg_general_pp_age=pd.DataFrame(avg_general_pp_age)
avg_general_pp_age.columns=["Avg Total Purchase per Person"]
avg_general_pp_age
```

```python
#Summary of purchasing analysis
Purchasing_analysis = age_range_general.copy() 
Purchasing_analysis["Average Purchase Price"] = avg_general_age["Average Purchase Price"]
Purchasing_analysis["Total Purchase Value"] = sum_general_age["Total Purchase Value"]
Purchasing_analysis["Avg Total Purchase per Person"] = avg_general_pp_age["Avg Total Purchase per Person"]                                              
Purchasing_analysis = Purchasing_analysis.reset_index()
Purchasing_analysis
```

```python
#Top Spenders
#Sort to see top 5
sn=purchase_data.groupby(["SN"])
sn_sum = sn["Price"].sum()
sn_sum_total=pd.DataFrame(sn_sum)
ordenar_top = sn_sum_total.sort_values("Price",ascending=False)
ordenar_top.columns = ["Top 5"]
ordenar_top.head()
```

```python
#Purchase count top
sn=purchase_data.groupby(["SN"])
sn_count_general=sn["Purchase ID"].count()
sn_general=pd.DataFrame(sn_count_general)
sn_general.columns=["Purchase Count"]
sn_general.head()
```

```python
#average purchase top
sn=purchase_data.groupby(["SN"])
sn_avg=round((sn["Price"].sum())/sn["Purchase ID"].count(),2)
sn_avg_general=pd.DataFrame(sn_avg)
sn_avg_general.columns=["Average Purchase Price"]
sn_avg_general.head()
```

```python
#purchase value general
sn=purchase_data.groupby(["SN"])
sum_general_sn=round((sn["Price"].sum()),2)
sum_general_sn=pd.DataFrame(sum_general_sn)
sum_general_sn.columns=["Total Purchase Value"]
sum_general_sn.head()
```

```python
#Summary of top
purchase_top = ordenar_top.copy() 
purchase_top["Purchase Count"] = sn_general["Purchase Count"]
purchase_top["Average Purchase Price"] = sn_avg_general["Average Purchase Price"]
purchase_top["Total Purchase Value"] = sum_general_sn["Total Purchase Value"]
purchase_top.head()
```

```python
#most popular items - purchase count
sn_=purchase_data.groupby(["Item ID","Item Name"])
count_sn=sn_["Item ID"].count()
sn_ct=pd.DataFrame(count_sn)
sn_ct.columns=["Purchase Count"]
ordenar_ = sn_ct.sort_values("Purchase Count",ascending=False)
ordenar_.columns = ["Purchase Count"]
ordenar_.head()
```

```python
#Summary of most popular items
purchase_= ordenar_.copy() 
purchase_["Item Price"] = sn_ct_price["Item Price"]
purchase_["Total Purchase Value"] = sn_ct_price_tt["Total Purchase Value"]
purchase_.head()
```
